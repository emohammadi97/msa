```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def tf_idf_cal(docs):
  vectorizer = TfidfVectorizer()
  doc_term_mat = vectorizer.fit_transform(docs)
  words = vectorizer.get_feature_names_out()
  doc_term_df = pd.DataFrame(doc_term_mat.toarray() , columns=words)
  n_doc, n_word = doc_term_df.shape
  return vectorizer, doc_term_mat, words, doc_term_df, n_doc, n_word

vectorizer, doc_term_mat, words, doc_term_df, n_doc, n_word = tf_idf_cal(s1)

distances = np.zeros((n_word, n_word))
for i in range(n_word):
    for j in range(i):
        cosineSimilarities = np.array(cosine_similarity(doc_term_mat[:,i].reshape(1,-1), doc_term_mat[:,j].reshape(1,-1)))
        distances[i, j] = cosineSimilarities
        distances[j, i] = distances[i, j]

df_distances = pd.DataFrame(1 - distances, index=[words], columns=[words])
```

```python
df_distances.values[[np.arange(df_distances.shape[0])]*2] = 1
```

```python
df_distances.loc['خدا','الله'] = .0040401
df_distances.loc['الله','خدا'] = .0040401

df_distances.loc['خدا','بسم'] = .0000031
df_distances.loc['بسم','خدا'] = .0000031

df_distances.loc['است','نیست'] = 0.0000005
df_distances.loc['نیست','است'] = 0.0000005

scoring_fn = lambda a, b: (1 - (df_distances.loc[a,b].iloc[0][0]))
```
